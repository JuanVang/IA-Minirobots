```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
curules = []
for x in range(50):
  curules.append(random.randint(1,5))#reparte las curules entre los partidos

representacion = []
for x in range(5):
  representacion.append(curules.count(x+1)/50)#el valor de la representacion de cada partido

peso_politico = []
for x in range(50):
  peso_politico.append(random.randint(1,100))

print(peso_politico)

```

```python
def representacion_peso(distribucion,partidos,peso_politico):
  rep_distribucion = [0]*partidos #Lista vacia de distribucion
  for y in range(partidos):
    for x in range(len(peso_politico)):
      if distribucion[x]==y+1:#Si el partido fue selecciono para el puesto
        to_add  = peso_politico[x] #Suma el peso del puesto
        rep_distribucion[y]+=to_add
  #print(rep_distribucion)
  for y in range(len(rep_distribucion)):
    rep_distribucion[y] = rep_distribucion[y]/sum(peso_politico)
  return rep_distribucion

def aptitud(distribucion,partidos,representacion):
  peso_total = sum(distribucion)
  aptitud_lista = [0]*5
  representacion_p = representacion_peso(distribucion,partidos,peso_politico)
  for x in range(5):
    aptitud_lista[x] = (abs(representacion_p[x]-representacion[x]))
  #print(aptitud_lista)
  return max(aptitud_lista)*2 - np.median(aptitud_lista)

def genera(K,l):
    pob = [np.random.randint(1, 6, l).tolist() for _ in range(K)]
    return pob
```

```python
def seleccion(pob):
  aptitud_lista = []
  for x in range(len(pob)):
    aptitud_lista.append(aptitud(pob[x],5,representacion))
  probabilidad = []
  for x in range(len(pob)):
    probabilidad.append(aptitud_lista[x]/sum(aptitud_lista))
  #print(probabilidad)
  j = 0
  K = len(pob)
  pob_nueva = []
  while j < K:
      i = 0
      while i < K:
          aleat = random.random()/2
          if probabilidad[i] > aleat:
              pob_nueva.append(pob[j])
              j += 1
              if j >= K:
                  break
          i +=1
          if i == K:
              i = 0
  return pob_nueva
```

```python
def cross(parents):
  hijos = []
  for x in range(0,len(parents),2):
    crom_1 = parents[x]
    if x>=len(parents)-1:
      crom_2 = parents[0]
    else:
      crom_2 = parents[x+1]
    punto = random.randint(1,len(crom_1)-1)
    punto2 = random.randint(1,len(crom_1)-1)
    if punto < punto2:
      hijo1 = crom_1[:punto] + crom_2[punto:punto2] + crom_1[punto2:]
      hijo2 = crom_2[:punto] + crom_1[punto:punto2] + crom_2[punto2:]
      hijos.append(hijo1)
      hijos.append(hijo2)
    elif punto > punto2:
      hijo1 = crom_1[:punto2] + crom_2[punto2:punto] + crom_1[punto:]
      hijo2 = crom_2[:punto2] + crom_1[punto2:punto] + crom_2[punto:]
      hijos.append(hijo1)
      hijos.append(hijo2)
    else:
      hijo1 = crom_1[:punto] + crom_2[punto:]
      hijo2 = crom_2[:punto] + crom_1[punto:]
      hijos.append(hijo1)
      hijos.append(hijo2)
  return hijos
```

```python
def mutation(hijos):
  for x in range(len(hijos)):
    if random.random() < 0.1:
      for y in range(len(hijos[x])):
        if random.random() < 0.1:
          punto = random.randint(0,len(hijos[x])-1)
          hijos[x][punto] = random.randint(1,5)

  return hijos
```

```python
def algoritmo_genetico(K):
  poblacion = genera(10,50)
  parents = seleccion(poblacion)
  print(parents[0])
  print(representacion_peso(parents[0],5,peso_politico))
  for x in range(K):
    hijos = cross(parents)
    hijos = mutation(hijos)
    parents = seleccion(hijos)
    if x%50 == 0:
      print("gen: ",x,", aptitud: ",aptitud(parents[0],5,representacion))
      print(len(hijos))
  return parents
```
